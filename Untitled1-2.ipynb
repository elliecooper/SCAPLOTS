```python
import pandas as pd
import numpy as np 
#%matplotlib inline
#%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
x_random = np.random.rand(100)*100  #random number generator
y_random = np.random.rand(100)*100
```

```python
fig = plt.figure(figsize=[10, 8]) #creating our figure
ax = fig.add_subplot(111, facecolor='Black')
ax.scatter(x_random, y_random, marker='*', color ='w') #creating a scatter plot
plt.savefig('Scatterplot_random.png')
```

```python
plt.show()
```
